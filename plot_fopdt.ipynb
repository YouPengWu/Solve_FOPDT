# Import packages

```python
import matplotlib.pyplot as plot
import numpy as np
from ipywidgets import interact
```

## FOPDT formula

$f(x) = 1 - e^{-(t-\theta) / \tau}$

In the formula, there are two constant to be trained: $\theta$ and $\tau$.

To train, we define a python function of $t$, $\theta$ and $\tau$.

```python
def fopdt(t, theta, tau):
    return 1 - np.exp(-(t - theta) / tau)
```

Normally, FOPDT output will need to multiply with another constant $K$ to convert to the right range.

```python
def fopdt_k(k, t, theta, tau):
    return k * fopdt(t, theta, tau)
```

plot the function

```python
def draw_fopdt(k = -0.38, theta=0, tau=220.1):
    t = np.linspace(0, 1000, 50)
    plot.figure()
    plot.xlabel("X")
    plot.ylabel("Y")
    plot.grid()
    plot.plot(t, fopdt_k(k, t, theta, tau))
    plot.show()

interact(draw_fopdt, k = (-1000, 1000, 0.01), theta=(-999,999,0.5), tau=(-999,999,0.5))
```
