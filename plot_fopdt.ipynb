```python
import matplotlib.pyplot as plot
from matplotlib.widgets import Slider
import numpy as np
from ipywidgets import interact

def fopdt(t, theta, tau):
    return 1 - np.exp(-(t - theta) / tau)

def fopdt_k(k, t, theta, tau):
    return k * fopdt(t, theta, tau)
```

```python
def draw_fopdt(k = -0.38, theta=0, tau=220.1):
    t = np.linspace(0, 1000, 50)
    plot.figure()
    plot.xlabel("X")
    plot.ylabel("Y")
    plot.grid()
    plot.plot(t, fopdt_k(k, t, theta, tau))
    plot.show()

interact(draw_fopdt, k = (-1000, 1000, 0.01), theta=(-999,999,0.5), tau=(-999,999,0.5))
```
